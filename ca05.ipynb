```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
```

```python
## Load Dataset
#The dataset is
#sourced from a GitHub repository provided in the assignment instructions. It contains movie features such as IMDB rating and genre indicators.
```

```python
url = "https://github.com/ArinB/MSBA-CA-Data/raw/main/CA05/movies_recommendation_data.csv"
df = pd.read_csv(url)
df.head()
```

## Data Preprocessing

We remove non-numerical and irrelevant columns such as:
- Movie ID
- Movie Name
- Label (not used for similarity)

This ensures the model only uses numerical features for distance calculations.

```python
df_clean = df.drop(columns=["Movie ID", "Movie Name", "Label"])
df_clean.head()
```

## Feature Matrix

We define the feature matrix using IMDB rating and genre columns. Each movie is represented as a numerical vector.

```python
X = df_clean
```

## Build kNN Model

We use the NearestNeighbors algorithm from scikit-learn with:
- k = 5 (number of neighbors)
- Euclidean distance as the similarity metric

```python
model = NearestNeighbors(n_neighbors=5, metric='euclidean')
model.fit(X)
```

## Query Movie: "The Post"

Feature values:
- IMDB Rating = 7.2  
- Biography = 1  
- Drama = 1  
- Thriller = 0  
- Comedy = 0  
- Crime = 0  
- Mystery = 0  
- History = 1  

```python
the_post = [[7.2, 1, 1, 0, 0, 0, 0, 1]]
```

## Find Similar Movies

We compute the 5 nearest neighbors using the trained kNN model but again it doesnt learn a pattern simply memorizes the dataset.

```python
distances, indices = model.kneighbors(the_post)

distances, indices
```

## Recommended Movies

Bellow are the 5 most similar movies to "The Post":

```python
recommended_movies = df.iloc[indices[0]][["Movie Name", "IMDB Rating"]]
recommended_movies
```

## Conclusion

The kNN algorithm successfully identifies movies similar to "The Post" based on genre and IMDB rating.

### Key Insights:
- Movies with similar genre combinations (Biography + Drama) are prioritized.
- IMDB rating also influences similarity.
- The model does not account for actors, directors, or storyline, which limits recommendation depth.

### Limitations:
- Small dataset (30 movies)
- Limited feature set
- No personalization or user preferences
- No learning involved, the model just memorizes the dataset

Despite these limitations, the model demonstrates how kNN can be applied to recommendation systems, this is a great example to show to companies in interviews as a sample of your work or even what can be achieved with their data.
